```python
mincraft_color_hex = ['E9ECEC',
'F07613',
'BD44B3',
'3AAFD9',
'F8C627',
'70B919',
'ED8DAC',
'3E4447',
'8E8E86',
'158991',
'792AAC',
'35399D',
'724728',
'546D1B',
'A12722',
'141519']
global mincraft_color_rgb
mincraft_color_rgb = []
for col in mincraft_color_hex:
  tup = tuple( int(col[x:x+2],16) for x in range(0,6,2))
  mincraft_color_rgb.append(tup)
mincraft_color_rgb


def choose_color(pixel):
    dists = np.linalg.norm( np.array([pixel] * len(mincraft_color_rgb)) - np.array(mincraft_color_rgb) )
    id = dists.index(min(dists))
    return mincraft_color_rgb[id]
```
